```python
from pinkylib import Camera
```

```python
cam = Camera()
```

```python
cam.set_calibration() #카메라 캘리브레이션 설정
```

```python
cam.start()
```

```python
frame = cam.get_frame()
pose = cam.detect_aruco(frame, marker_size=0.036)
print("x: ", str(pose[0]), "y: ", str(pose[1]), "z: ", str(pose[2]))
```

```python
cam.close()
```
